```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the CSV file
res_rating = pd.read_csv("/content/Dataset.csv",delimiter=",")
res_rating
```

```python
# Check for missing values
print("Missing values before handling:")
print(res_rating.isnull().sum())

categorical_cols = ['Restaurant Name', 'City', 'Address', 'Locality', 'Locality Verbose', 'Cuisines', 'Currency',
                    'Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu',
                    'Rating color', 'Rating text']

# Apply one-hot encoding to the identified categorical columns
res_rating_encoded = pd.get_dummies(res_rating, columns=categorical_cols, dummy_na=False)

# Display the first few rows and information about the processed DataFrame
print("\n")
print("Processed DataFrame after encoding:")
display(res_rating_encoded.head())
print("\n")
print("Processed DataFrame Info:")
res_rating_encoded.info()
```

```python
# Let's re-do the encoding after filling missing values in the original 'Cuisines' column.
res_rating['Cuisines'].fillna('Unknown', inplace=True)

# Now, re-apply one-hot encoding to the identified categorical columns including the handled 'Cuisines'
categorical_cols = ['Restaurant Name', 'City', 'Address', 'Locality', 'Locality Verbose', 'Cuisines', 'Currency',
                    'Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu',
                    'Rating color', 'Rating text']

res_rating_encoded = pd.get_dummies(res_rating, columns=categorical_cols, dummy_na=False)

# Display the first few rows and information about the processed DataFrame to confirm the data is ready
print("\nProcessed DataFrame after handling missing values and re-encoding:")
display(res_rating_encoded.head())
print("\nProcessed DataFrame Info:")
res_rating_encoded.info()
```

```python
# Examine the columns of the res_rating_encoded DataFrame.
print("Columns in res_rating_encoded:")
print(res_rating_encoded.columns)
recommendation_features = []

# Add columns related to Cuisines
cuisine_cols = [col for col in res_rating_encoded.columns if col.startswith('Cuisines_')]
recommendation_features.extend(cuisine_cols)
recommendation_features = []

# Cuisines (one-hot encoded)
cuisine_cols = [col for col in res_rating_encoded.columns if col.startswith('Cuisines_')]
recommendation_features.extend(cuisine_cols)

# Based on the previous successful run for data preparation, 'Price range' was NOT one-hot encoded.
# It seems 'Price range' was kept as a numerical column in the final res_rating_encoded.
if 'Price range' in res_rating_encoded.columns:
    recommendation_features.append('Price range')

# Average Cost for two (numerical)
if 'Average Cost for two' in res_rating_encoded.columns:
    recommendation_features.append('Average Cost for two')

# Aggregate rating (numerical)
if 'Aggregate rating' in res_rating_encoded.columns:
    recommendation_features.append('Aggregate rating')

# Votes (numerical)
if 'Votes' in res_rating_encoded.columns:
    recommendation_features.append('Votes')

# Rating text (one-hot encoded)
rating_text_cols = [col for col in res_rating_encoded.columns if col.startswith('Rating text_')]
recommendation_features.extend(rating_text_cols)

# Rating color (one-hot encoded)
rating_color_cols = [col for col in res_rating_encoded.columns if col.startswith('Rating color_')]
recommendation_features.extend(rating_color_cols)

# Display the list of selected feature names
print("\nSelected features for recommendation criteria:")
print(recommendation_features)

# Create a DataFrame with only the selected features
features_df = res_rating_encoded[recommendation_features]

# Display the first few rows of the features DataFrame
print("\nDataFrame with selected features:")
display(features_df.head())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate the cosine similarity matrix
cosine_sim = cosine_similarity(features_df)

# Print the shape of the similarity matrix
print("Shape of the cosine similarity matrix:", cosine_sim.shape)
```

```python
def get_recommendations(restaurant_name, cosine_sim=cosine_sim, original_df=res_rating, encoded_df=res_rating_encoded, num_recommendations=10):

    # Get the index of the restaurant that matches the name from the original DataFrame
    if restaurant_name not in original_df['Restaurant Name'].values:
        print(f"Restaurant '{restaurant_name}' not found.")
        return []

    # Use the index from the original DataFrame to get the corresponding row in the encoded DataFrame
    idx = original_df[original_df['Restaurant Name'] == restaurant_name].index[0]

    # Get the similarity scores for this restaurant from the cosine similarity matrix
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the similarity scores in descending order
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the indices of the top num_recommendations most similar restaurants (excluding the restaurant itself)
    top_indices = [i[0] for i in sim_scores[1:num_recommendations + 1]]

    # Use these indices to retrieve the names of the recommended restaurants from the original DataFrame
    recommended_restaurant_names = original_df['Restaurant Name'].iloc[top_indices].tolist()

    return recommended_restaurant_names

# Test the function with a sample restaurant name
sample_restaurant_name = 'Le Petit Souffle'
recommendations = get_recommendations(sample_restaurant_name)

print(f"\nRecommendations for '{sample_restaurant_name}':")
print(recommendations)
```
