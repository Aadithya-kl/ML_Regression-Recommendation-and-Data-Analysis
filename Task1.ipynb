```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the CSV file
res_rating = pd.read_csv("/content/Dataset.csv",delimiter=",")
res_rating
```

```python
res_rating.shape
```

```python
res_rating.info() #This gives the information about the columns of dataset, Using this we find all the non-numerical columns in dataset
#The given dataset doesn't have any null values.
```

```python
res_rating.describe()
```

```python
# Converting all the non-numerical columns in dataset to numerical columns using Label Encoder
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()

# Select the columns to encode by name
encode_colm = res_rating[['Restaurant ID', 'Restaurant Name', 'City', 'Address', 'Locality', 'Locality Verbose', 'Latitude', 'Cuisines', 'Average Cost for two', 'Currency', 'Switch to order menu', 'Price range', 'Aggregate rating', 'Rating color', 'Rating text', 'Votes', 'Has Table booking', 'Has Online delivery', 'Is delivering now']]

for col in encode_colm.columns:
    res_rating[col] = label.fit_transform(res_rating[col])
res_rating
```

```python
# Correlation Matrix
correlation_Matrix = res_rating.corr()
correlation_Matrix
```

```python
# Scatter Matrix
pd.plotting.scatter_matrix(res_rating, figsize=(15, 10))
plt.show()
```

```python
# Storing the input attributes and the target variable
X = res_rating.drop('Aggregate rating', axis=1) # Input attributes
y = res_rating['Aggregate rating'] # Target variable
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
stand_scale = StandardScaler()
stand_scale.fit(X_train)
X_train_std = stand_scale.transform(X_train)
X_test_std = stand_scale.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
LR_model = LinearRegression()
LR_model.fit(X_train_std,y_train)
LR_score = LR_model.score(X_test_std,y_test) # Finding the R2 score of the model
print("R2 score:",LR_score)
```

```python
from sklearn.tree import DecisionTreeRegressor

Decision_model = DecisionTreeRegressor(random_state=42)
Decision_model.fit(X_train_std, y_train)
Decision_score = Decision_model.score(X_test_std, y_test)
print("Decision Tree R2 score:", Decision_score)
```

```python
# Evaluate the model's performance using appropriate regression metrics
from sklearn.metrics import mean_squared_error, r2_score

# For Linear Regression
y_pred_LR = LR_model.predict(X_test_std)
mse_LR = mean_squared_error(y_test, y_pred_LR)
rmse_LR = np.sqrt(mse_LR)
r2_LR = r2_score(y_test, y_pred_LR)

print("Linear Regression Metrics:")
print("Mean Squared Error:", mse_LR)
print("Root Mean Squared Error:", rmse_LR)
print("R-squared:", r2_LR)

# For Decision Tree Regressor
y_pred_DT = Decision_model.predict(X_test_std)
mse_DT = mean_squared_error(y_test, y_pred_DT)
rmse_DT = np.sqrt(mse_DT)
r2_DT = r2_score(y_test, y_pred_DT)
print("\n")
print("Decision Tree Regressor Metrics:")
print("Mean Squared Error:", mse_DT)
print("Root Mean Squared Error:", rmse_DT)
print("R-squared:", r2_DT)
```

```python
# Model's results and analyze the most influential features affecting restaurant ratings.
# For Linear Regression
print("Linear Regression Coefficients:")
for col, coef in zip(X.columns, LR_model.coef_):
    print(f"{col}: {coef}")

# For Decision Tree Regressor
print("\nDecision Tree Regressor Feature Importances:")
for col, importance in zip(X.columns, Decision_model.feature_importances_):
    print(f"{col}: {importance}")
```
